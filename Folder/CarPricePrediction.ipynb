Import all the necessary libraries needed for this project. Also, create a link to the dataset. The file that holds the information for this dataset is called 'train.csv'.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv('train.csv')
dataset
```

Now let's take a look at how the data is stored, what data types they are, as well as the null/nan values.

```python
dataset.shape
```

We can see that we have a total of 19106 records that contain 36 different attributes.

```python
dataset.isnull().sum()
```

There are a lot of null values from the above dataset. Since we have about 19000 records, we can consider dropping these values from our dataset, since however we try to replace them (by mean, mode, or otherwise), there will always be some unrealealistic replacement. This can allow the rest of the data to be more precise.

```python
dataset = dataset.dropna(how='any', axis=0)
dataset.shape
```

We can see that we dropped about 5000 records, and now we know that we have no null/nan values.

```python
dataset.isnull().sum()
```

Now let's see the datatypes of all these attributes.

```python
dataset.dtypes
```

From above, we can see that the datatypes seem to match with the corresponding attributes, but there is an odd one out. Price is stored as an object type, which should not be the case. It turns out that the creator of the dataset, whenever they did not find a corresponding value for the price of a car, they had put the label 'ot Priced', which in turn converts the entire price attribute to an object type. We have to remove these records, as well as convert the remainder of the attribute to an int type.

```python
dataset = dataset[~dataset['price'].isin(['ot Priced'])] # removes all tuples with 'ot Priced' label
dataset['price'] = dataset['price'].astype(int) # converts all objects to int
print(dataset.shape)
print(dataset['price'].dtypes)
```

Above we can see that the records that we mentioned are removed and the price attribute is converted to an integer data type. Now we can try to see the different attributes and the correlation between those attributes and the target variable, which will be the price.

```python
dataset.corr(numeric_only=True)
```

The correlation can only contrast integer or float datatypes, so in this example we lose important information that is stored under the attributest that are stored as object types, like brand, model, and so on. Most of the people base the price on the brand and model of a car, and if we cannot represent them, we loose a lot of information. To gain the information from them, we need to convert them to a compatible datatype, but first we need to see how the brands, models and so on, compare to the respective prices. For this comparison, we will compare the average price of each brand.

```python
brand_average = dataset.groupby('brand')['price'].mean()
brand_average.plot(kind='bar')
plt.show()
```

From the graph above, we can see the average price of each car brand, and as we can also see, there is also a brand category called Suzuki, which is not a car brand so we can take it out since it could affect our model based on cars.

```python
dataset = dataset.drop(dataset[dataset['brand'] == 'Suzuki'].index)
brand_average = dataset.groupby('brand')['price'].mean()
brand_average.plot(kind='bar')
plt.show()
```

As we can see, there is a correlation between the price and the brand of the car. Some brands sell on average for much cheaper, like FIATs, and some sell for much more on average, like Porsches. We can do a similar average calculation for all other object datatypes, and compare their respective correlation to the price attribute.

```python
model_average = dataset.groupby('model')['price'].mean()
engine_average = dataset.groupby('engine')['price'].mean()
transmission_average = dataset.groupby('transmission')['price'].mean()
fuel_type_average = dataset.groupby('fuel_type')['price'].mean()
drivetrain_average = dataset.groupby('drivetrain')['price'].mean()
interior_color_average = dataset.groupby('interior_color')['price'].mean()
exterior_color_average = dataset.groupby('exterior_color')['price'].mean()
```

Now that we have these values, we can merge them with our dataset, and call the correlation method and see how they compare.

```python
%%time
map_to_brand = brand_average.to_dict()
dataset['avg_brand'] = dataset['brand'].map(map_to_brand)

map_to_model = model_average.to_dict()
dataset['avg_model'] = dataset['model'].map(map_to_model)

map_to_engine = engine_average.to_dict()
dataset['avg_engine'] = dataset['engine'].map(map_to_engine)

map_to_transmission = transmission_average.to_dict()
dataset['avg_transmission'] = dataset['transmission'].map(map_to_transmission)

map_to_fuel_type = fuel_type_average.to_dict()
dataset['avg_fuel_type'] = dataset['fuel_type'].map(map_to_fuel_type)

map_to_drivetrain = drivetrain_average.to_dict()
dataset['avg_drivetrain'] = dataset['drivetrain'].map(map_to_drivetrain)

map_to_interior_color = interior_color_average.to_dict()
dataset['avg_interior_color'] = dataset['interior_color'].map(map_to_interior_color)

map_to_exterior_color = exterior_color_average.to_dict()
dataset['avg_exterior_color'] = dataset['exterior_color'].map(map_to_exterior_color)
```

We can use these models to create an accurate model, as well as remove other attributes that would not contribute as much to the model. Now lets see and remove as many outliers from our dataset as possible. To do this, lets do some comparisons between the year, mileage, and price.

```python
%%time
dataset.plot(x='mileage', y='price', kind='scatter')
plt.show()
```

From the graph above, we cans see that there are some extremely high prices, that could mislead our model. Also, there are some cars with a lot of miles, so we need to take care of them as well.

```python
%%time
dataset.nlargest(10, 'price')
dataset.nlargest(10, 'mileage')
```

We can set the maximum price to 300'000$ and the maximum mileage to 250'000 miles. This will remove those outliers that are way out of range compared to the rest of the dataset.

```python
condition1 = dataset['price'] <= 300000
condition2 = dataset['mileage'] <= 250000

dataset = dataset[condition1]
dataset = dataset[condition2]

dataset.plot(x='mileage', y='price', kind='scatter')
plt.show()
```

From the graph above, we can see that there is a negative correlation between the mileage and the price, and it represents a hyperbola, which might hinder possible machine learning algorithms we will use when modeling our algorithms. Next we will see how the production year of the car affects the price.

```python
dataset.plot(x='year', y='price', kind='scatter')
plt.show()
```

As we can see, as the newer the year the more the car costs, but there are some outliers from the older days, probably because they are considered as classical cars. We need to remove them from our dataset because they can misslead our models into valuing all old ages to be more expensive.

```python
condition3 = dataset['year'] >= 1998
dataset = dataset[condition3]

dataset.plot(x='year', y='price', kind='scatter')
plt.show()
```

Now we can see a positive correlation as we were expecting. With our dataset modified and adjusted, we can start playting with the ML models. We will be working with the KNN and Multilinear Regression algorithms.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import mean_squared_error

X = dataset[['year', 'mileage', 'navigation_system', 'min_mpg', 'max_mpg', 'memory_seat', 'leather_seats', 'apple_car_play/android_auto', 'avg_brand', 'avg_model', 'avg_engine', 'avg_transmission', 'avg_drivetrain', 'avg_interior_color', 'avg_exterior_color']]
y = dataset['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

def weight(distances):
    return 1/distances**2


model = KNeighborsClassifier(n_neighbors=4, weights=weight)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
msle = mean_squared_log_error(y_test, y_pred)
mae = (y_test-y_pred).abs().mean()

plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_pred), max(y_pred)], color='red', linestyle='--')
plt.xlabel('Real Values')
plt.ylabel('Predicted Values')
plt.title('K-Nearest Neighbours')
plt.grid(True)
plt.show()

print("MAE:", mae)
print("R^2:", r2)
print("MSLE:", msle)
```

As we mentioned for the mileage in the mileage section, it has a hyperbolic correltaion with price. Multilinear regression uses the formula y = b0 + b1 * x + b2 * x + ..., and when we add an input with a correlation different from the linear trajectory, it could misslead the model.

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import mean_squared_error

X = dataset[['year', 'mileage', 'navigation_system', 'min_mpg', 'max_mpg', 'memory_seat', 'leather_seats', 'apple_car_play/android_auto', 'avg_brand', 'avg_model', 'avg_engine', 'avg_transmission', 'avg_drivetrain', 'avg_interior_color', 'avg_exterior_color']]
y = dataset['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train)

y_pred_L_R = linear_regression_model.predict(X_test)

r2 = r2_score(y_test, y_pred_L_R)
mse = mean_squared_error(y_test, y_pred_L_R)
mae = (y_test - y_pred_L_R).abs().mean()

plt.scatter(y_test, y_pred_L_R, label = 'Upper', color = 'red')
plt.plot([min(y_test), max(y_test)], [min(y_pred_L_R), max(y_pred_L_R)], color='blue', linestyle='--')
plt.xlabel('Real Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression')
plt.grid(True)
plt.show()

print("R^2:", r2)
print("MAE:", mae)
```

Above, we can see that there is a small deviation at the start, and that could be due to what we described earlier, where one of the parameters could affect how the model might perform, but overall it generates a similar value to the KNN algorithm, which means that the models are relatively efficient.
